```python
from pymongo import MongoClient
import pandas as pd

# --- Cấu hình MongoDB ---
MONGO_URI = "mongodb://localhost:27017/"
DB_NAME = "riot_db"
COLLECTION_NAME = "participant"

# --- Kết nối MongoDB ---
client = MongoClient(MONGO_URI)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]

# --- Pipeline Aggregation ---
pipeline = [
    {
        "$group": {
            "_id": "$info.championName",
            
            # Tính trung bình và tối đa damage
            "average_damage_to_champions": {"$avg": "$info.totalDamageDealtToChampions"},
            "max_damage_to_champions": {"$max": "$info.totalDamageDealtToChampions"},
            
            # Tính trung bình và tối đa pentakill
            "average_pentakill": {"$avg": "$info.pentaKills"},
            "max_pentakill": {"$max": "$info.pentaKills"},
            
            # Tính số trận thắng và tổng số trận
            "wins": {
                "$sum": {
                    "$cond": [{"$eq": ["$info.win", True]}, 1, 0]
                }
            },
            "total_games": {"$sum": 1}
        }
    },
]

# --- Truy vấn dữ liệu ---
result = list(collection.aggregate(pipeline))

# Chuyển sang DataFrame
df = pd.DataFrame(result)
df = df.rename(columns={"_id": "Champion"})
```

```python
def get_info_champion(championName):
    champion_info = df.loc[df["Champion"] == championName, :]
    return champion_info
```

```python
champion_info = get_info_champion("Zed")
```

```python
def plot_winrate(champion_info):
    win_rate = round(float(champion_info["wins"] / champion_info["total_games"]), 2)
    percent = [win_rate, 1 - win_rate]
    plt.pie(percent, colors=["green", "red"], labels=["win", "lose"], autopct="%.f")
    plt.show()
```

```python
plot_winrate(champion_info)
```
